```python
 #calling the dependencies
from deepface import DeepFace
import cv2 as cv
import matplotlib.pyplot as plt 
```

```python
 #importing the images
img1_path = '/Users/bhurv/FaceRecognition/Images/adult.png'

 #confirming the path of images
img1 = cv.imread(img1_path)
```

```python
plt.imshow(img1[:, :, ::-1 ]) 
#setting value as -1 to maintain saturation
plt.show()
```

```python
#calling VGGFace
model_name = "VGG-Face"
model = DeepFace.build_model(model_name) 
```

```python

obj = DeepFace.analyze(img_path = img1, actions = ['age', 'gender', 'race', 'emotion'])
print(int(obj["age"])," years old ",obj["dominant_race"]," ",obj["dominant_emotion"]," ", obj["gender"])
```
